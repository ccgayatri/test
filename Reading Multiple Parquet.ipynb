```python
import zipfile
import pyarrow.parquet as pq
import pandas as pd
import io

# Path to your zip file in Google Drive
zip_file_path = '/content/drive/MyDrive/CC_GAYATRI/OneDrive_1_4-25-2024 (1).zip'

# Open the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    file_list = zip_ref.namelist()

    # Iterate over each file in the zip archive
    for file_name in file_list:
        if file_name.endswith('.parquet'):
            # Read the Parquet file from the zip archive into memory
            with zip_ref.open(file_name) as file:
                parquet_file = io.BytesIO(file.read())

                # Read the Parquet file using pyarrow
                table = pq.read_table(parquet_file)

                # Convert to pandas DataFrame (optional)
                df = table.to_pandas()

                # Now you can work with the DataFrame (df) or the pyarrow Table (table)
                print(f"Contents of {file_name}:")
                print(df.head())  # Display first few rows of the DataFrame
```

```python
import os
import pandas as pd
import pyarrow.parquet as pq

# Path to the parent folder containing subfolders with Parquet files
parent_folder_path = '/content/drive/MyDrive/CC_GAYATRI/Mar.parquet'

# Initialize an empty list to store DataFrame chunks
dfs = []

# Iterate through each subfolder (and its contents) in the parent folder
for root, dirs, files in os.walk(parent_folder_path):
    for file in files:
        if file.endswith('.parquet'):
            # Construct the full path to the Parquet file
            parquet_file_path = os.path.join(root, file)

            # Read the Parquet file using pyarrow
            table = pq.read_table(parquet_file_path)

            # Convert to pandas DataFrame
            df_chunk = table.to_pandas()

            # Append the DataFrame chunk to the list
            dfs.append(df_chunk)

# Concatenate all DataFrame chunks into a single DataFrame
final_df = pd.concat(dfs, ignore_index=True)

# Display the final DataFrame
print("Final DataFrame:")
print(final_df.head())
```

```python
import os
import pandas as pd
import pyarrow.parquet as pq

# List of parent folder paths
parent_folder_paths = [
    '/content/drive/MyDrive/CC_GAYATRI/Jan.parquet',
    '/content/drive/MyDrive/CC_GAYATRI/Feb.parquet',
    '/content/drive/MyDrive/CC_GAYATRI/Mar.parquet'
]

# Initializing an empty list to store DataFrame chunks
dfs = []

# Iterate through each parent folder
for parent_folder_path in parent_folder_paths:
    # Iterate through each subfolder and content in the parent folder
    for root, dirs, files in os.walk(parent_folder_path):
        for file in files:
            if file.endswith('.parquet'):

                parquet_file_path = os.path.join(root, file)
                table = pq.read_table(parquet_file_path)

                # Convert to pandas DataFrame
                df_chunk = table.to_pandas()
                # Append the DataFrame chunk to the list
                dfs.append(df_chunk)

# Concatenate all DataFrame chunks into a single DataFrame
final_df = pd.concat(dfs, ignore_index=True)

# Display the final DataFrame
print("Final DataFrame:")
print(final_df.head())
```
